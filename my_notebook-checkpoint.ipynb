<h1>Rule mining</h1>

<h3>Creating new df only with order_id and product_name</h3>

```python
#Reason:    No need of other columns

#Action:    created new aprior_df 

aprior_df = combined_df[['order_id','product_name']].copy()
aprior_df.rename(columns={'order_id':'order','product_name':'items'},inplace=True)
aprior_df['marker']=1
```

<h3>Changing data type</h3>

```python
#Reason:    columns with id are expected to be object and not numerical.

#Action:    changed id's dtype numerical to object.

for col in aprior_df.columns:
    if ("id" in col) & (aprior_df[col].dtype != "object"):
        aprior_df[col] = aprior_df[col].astype(str)

# Confirm changes
aprior_df.info()
```

<h3>Preparing Invoice-Product Matrix for ARL Data Structure</h3>

```python
def make_matrix(aprior_df):
        return aprior_df.groupby(['order','items'])['marker'].sum().unstack().fillna(0). \
            applymap(lambda x: 1 if x > 0 else 0)
```

```python
df_matrix = make_matrix(aprior_df)
```

```python
df_matrix
```

```python
df_matrix.dtypes
```

<h3>Determination of Association Rules Frequent Itemsets Mining</h3>

```python
freq_itemsets = apriori(df_matrix, min_support=0.01, use_colnames=True)

# Collecting the inferred rules in a dataframe 
rules = association_rules(freq_itemsets,metric='support',min_threshold=0.01)
rules.sort_values('support',ascending=False)
```

```python
# Let's sort dataframe by lift:
rules.sort_values("lift", ascending=False)
```

```python
rules.sort_values("confidence", ascending=False)
```

```python
# Another rule where support more than 0.009 for given itemset
filtered_rules = rules[(rules['antecedent support'] > 0.01)&
                        (rules['consequent support'] >0.01) &
                        (rules['confidence'] > 0.2) &
                        (rules['lift'] > 1.0)&
                        (rules['support']>0.009)]
```

```python
filtered_rules
```

```python
def rules_to_coordinates(rules):
    rules['antecedent'] = rules['antecedents'].apply(lambda antecedent:list(antecedent)[0])
    rules['consequent'] = rules['consequents'].apply(lambda consequent:list(consequent)[0])
    rules['rule'] = rules.index
    return rules[['antecedent','consequent','rule']]
```

```python
# Convert rules into coordinates suitable for use in a parallel coordinates plot
coords = rules_to_coordinates(rules)
# Generate parallel coordinates plot
plt.figure(figsize=(3,6))
parallel_coordinates(coords, 'rule',colormap = 'ocean')
plt.legend([])
plt.show()
```

<h1>Suggesting a Product to Users at the Basket Stage</h1>

```python


def check_product(item, df =aprior_df):
    product_name =  df[df["items"] == item]["items"].unique()[0]
    return product_name

```

```python
def recommend_product(item, df = aprior_df):
    products = item
    code_list = []
    name_list = []
    sorted_rules = rules.sort_values("lift", ascending=False)
    for idx, product in enumerate(sorted_rules["antecedents"]):
        for j in list(product):
            if j == products:
                code_list.append(list(sorted_rules.iloc[idx]["consequents"])[0])
                code_list = list(dict.fromkeys(code_list))
                
        # Print product name for the df 
    if len(code_list) <= 0:
        return "Invalid Product, try again!"
    else:
        for code in code_list:
            name_list.append(check_product(code))
    
    
    print(f"Recommended product(s) for {item} is/are:")
    for i, item in enumerate(name_list):
        print(i+1, item)

```
